# Backward experiment on sinthetics data

```python
import sys, os
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('../modules')
from nsw.nsw_classifier import NSWClassifier
import nsw.data_gen as dg
import random
from numpy.linalg import norm as vector_norm
```

## Prepare data and a cut

```python
G = NSWClassifier()

N = 4000
values = []
for i in range(N):
    p = np.array([random.random(), random.random()])
    cls = 1 if vector_norm(p - np.array([.5, .5])) < 0.4 else 0
    if random.random() > .97: cls = 1 - cls
    values.append((p, cls))

G.build_navigable_graph(values, attempts=10)

rich_cut = sorted([(e, G.dist(G.nodes[e[0]].value, G.nodes[e[1]].value)) for e in G.cut], key=lambda v: v[1])
lengths = [e[1] for e in rich_cut]
plt.hist(lengths, bins=100)
plt.show()
```

```python
smaller_cut = rich_cut[:2000]
print(smaller_cut[-1])
```

## Prepare ray-casting classifier

```python
from nsw.cut_classifier import CutClassifier
cclf = CutClassifier(G, smaller_cut)
clf = cclf.classifier
```

## Prepare test set

```python
N = 400
test = []
for i in range(N):
    p = np.array([random.random(), random.random()])
    cls = 1 if vector_norm(p - np.array([.5, .5])) < 0.4 else 0
    test.append((p, cls))
```

## Compare classifier with best hyperparameter sets

```python
print("K\tCut\tSupport\tFull")
for KK in [5, 9, 13]:
    for small, R, k, M in [
#         [0.005, 4, 5, 8], 
#         [0.005, 2, 9, 8], 
        [0.005, 2, 13, 8], 
#         [0.005, 1, 5, 32], 
#         [0.1, 4, 9, 16]
    ]:

        fullknn = 0
        for t in test:
            # simple kNN classification using FULL set
            nodes = G.multi_search(t[0], attempts=1, top=KK)
            classes = [G.nodes[n]._class for n in nodes]
            pc = sum(classes) > KK // 2
            if pc == t[1]:
                fullknn += 1


        continuous = 0
        for t in test:
            pc = clf(t[0], R=R, small=small, closest=k, M=M)
            if int(pc > .5) == t[1]:
                continuous += 1

        backward = 0
        for t in test:
            # simple kNN classification using support set
            nodes = cclf.support_nsw.multi_search(t[0], attempts=1, top=KK)
            classes = [cclf.support_nsw.nodes[n]._class for n in nodes]
            pc = sum(classes) > KK // 2
            if pc == t[1]:
                backward += 1

        print(f"{KK}\t{continuous / len(test)}\t{backward / len(test)}\t{fullknn / len(test)}")
```
